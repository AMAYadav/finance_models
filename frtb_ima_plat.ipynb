```python
class reft(valuations):

  import pandas as pd
  import numpy as np
  from scipy.stats import norm , spearmanr , ks_2samp
  import datetime

  def __init__(self,
               dates=None,
               share_price:float=None,
               option_price:float=None,
               position:str=['call','put'],
               rf:float=None,
               time_to_maturity= None,
               expiry=None,
               bsm_vol=None,
               delta=None,
               gamma=None,
               vega=None,
               theta=None,
               pnl=None
               ):
    super().__init__()

    self.delta = delta
    self.gamma=gamma
    self.vega=vega
    self.theta=theta
    self.pnl = pnl
    self.time_to_maturity = time_to_maturity

    if isinstance(dates,datetime.date):
      self.dates=dates
    else:
      self.dates=pd.to_datetime(dates) # dt.strftime()###--------------###
      # raise TypeError('dates object is not datetime')

    if isinstance(expiry, datetime.date):
      self.expiry=expiry
    else:
      self.expiry=pd.to_datetime(expiry)
# ---
    if isinstance(share_price, float):
      self.share_price=share_price
    else:
      self.share_price=(share_price).astype(float)
      # raise TypeError('share_price object is float')

    if isinstance(bsm_vol, float):
      self.bsm_vol=bsm_vol
    else:
      self.bsm_vol= (bsm_vol).astype(float)

    if isinstance(rf, float):
      self.rf=rf
    else:
      self.rf= (rf).astype(float)
    # else:
    #   raise TypeError('rf object is float')

    if isinstance(option_price, float):
      self.option_price=option_price
    else:
      self.option_price=(option_price).astype(float)
    # else:
    #   raise TypeError('option_price object is float')

# ------------------------------------------------------------------------------
  def time_to_maturity_1(self,y=250):
      start = np.array(self.dates).astype('datetime64[D]')
      expiry = np.array(self.expiry).astype('datetime64[D]')
      return  pd.DataFrame({'time_to_maturity':np.busday_count(start,expiry) / y,
                            'time_to_maturity': self.dates})
# ------------------------------------------------------------------------------
  def apl_2_0(self,):
    option_price=np.array(self.option_price)
    array = np.roll(np.diff(option_price),axis=0,shift=1)
    array = np.append(array, 0)
    return pd.DataFrame({'apl': array},index = self.dates)
    pass
# ------------------------------------------------------------------------------
  def hpl_2_1(self,): # bs_vol,delta,gamma,vega,theta
    ##since it is in dict format
    option_price=np.array(self.option_price)
    bs_vol=np.array(self.bsm_vol)
    delta=self.delta.iloc[1:]       ###write formula or take values
    gamma=self.gamma.iloc[1:]       ###write formula or take values
    vega=self.vega.iloc[1:]         ###write formula or take values
    theta=self.theta.iloc[1:]       ###write formula or take values

    time_to_maturity = self.time_to_maturity   #time_to_maturity_1().iloc[1:].values

    hpl=delta * np.diff(option_price) + 0.5 * gamma * np.diff(option_price) + vega * np.diff(bs_vol) + theta * time_to_maturity
    return pd.DataFrame({'hpl': hpl }).fillna(0).round(3).set_index([self.dates])
    pass
# ------------------------------------------------------------------------------
  def rtpl_2_2(self,):
    option_price=np.array(self.option_price)
    delta = np.array(self.delta.iloc[1])
             ####formula for delta ############
    array = np.roll( np.diff( option_price ) * delta    ,axis=0,shift=1)
    array = np.append(array, 0)
    return pd.DataFrame({'rtpl':array},index = self.dates)
# ------------------------------------------------------------------------------
  def var_2_3(self,confidence_level = 95):
                                 # self.delta  write your own formula # Calculate VaR
   value_at_risk = self.option_price * self.delta * self.bsm_vol * np.percentile(self.option_price , 100 - confidence_level)
   return pd.DataFrame({'var':value_at_risk}).set_index([self.dates])
  pass
# ------------------------------------------------------------------------------
  def exceptions_3(self,apl,hpl,var):

    if not isinstance(apl, pd.DataFrame):
      raise TypeError(f" {apl} is not in pandas dataframe ")

    if not isinstance(hpl, pd.DataFrame):
      raise TypeError(f" {hpl} is not in pandas dataframe ")

    if not isinstance(var, pd.DataFrame):
      raise TypeError(f" {var} is not in pandas dataframe ")
# ------------------------------------------------------------------------------
    #apl
    apl_exception = np.where(apl.values < var.values , 1 , 0)
    hpl_exception=  np.where(hpl.values < var.values , 1 , 0)

# ------------------------------------------------------------------------------
  #total exceptions
    e = np.maximum(np.sum(apl_exception),np.sum(hpl_exception))
    if (e>0 and e<=5):
      print (f'total exceptions: {e} ,green zone ')
      print (f'')
      print (f'model is accurate')
    elif(e>5 and e<=8):
      print(f'total exceptions: {e} ,amber zone ')
      print(f'')
      print(f'model disallowed')
    else:
      print(f'total exceptions: {e} ,red zone ')
      print(f'')
      print(f'model suspended in use')
    print()
    return pd.DataFrame({
        'total_exceptios':[ e ],
            'apl_exceptios':[np.sum(apl_exception)],
            'hpl_exceptios':[np.sum(hpl_exception)]})
    pass

  def stats_4(self,hpl,rtpl):

  # Calculate Spearman correlation

   corr, p_value = spearmanr(hpl,rtpl)

   print("Spearman correlation coefficient:", corr.round(5))
   print()
   print("P-value:", p_value.round(5))

  #  diff = ks_2samp([hpl.values ,self.pnl], [rtpl.values ,self.pnl])

  #  return diff
  pass
```
