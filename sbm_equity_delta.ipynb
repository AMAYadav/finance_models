```python
##main 2

at=d.pivot_table(index=['bucket',],columns=['risk_factor_id'],values='wsk',aggfunc='sum',margins=False,margins_name='wsk')
at.fillna(0,inplace=True)
at
```

```python
##main 3
alpha=d[['bucket','risk_factor_id','risk_weights','wsk','correl_base','correl_high','correl_low']]
alpha
```

```python
## main 4
pivot=alpha.groupby(['bucket','correl_base','correl_high','correl_low']).agg({'wsk':'sum',})
pivot
```

```python
## main 5
pivot=pivot.reset_index(['correl_base', 'correl_high', 'correl_low'])
pivot
```

```python
## main 6
ad= alpha.groupby(['bucket','correl_base','correl_high','correl_low']).agg({'wsk':'sum',})
ad['kb_correl_base']=np.sqrt(np.matmul(np.matmul(pivot.correl_base,at),(at.T).values))
ad['kb_correl_high']=np.sqrt(np.matmul(np.matmul(pivot.correl_high,at),(at.T).values))
ad['kb_correl_low']=np.sqrt(np.matmul(np.matmul(pivot.correl_low,at),(at.T).values))
ad.head(1)
```

```python
ad['sb_alt_base']=np.maximum(np.minimum(ad.wsk,ad.kb_correl_base),-ad.kb_correl_base).values
ad['sb_alt_high']=np.maximum(np.minimum(ad.wsk,ad.kb_correl_high),-ad.kb_correl_high).values
ad['sb_alt_low']=np.maximum(np.minimum(ad.wsk,ad.kb_correl_low),-ad.kb_correl_low).values
ad.round(4)
```

```python
correl_inter_high=np.minimum(100/100,1.25*correl_inter_base)*100
correl_inter_high
```

```python
correl_inter_low=np.maximum(2*correl_inter_base-100/100,(75/100)*correl_inter_base)*100
correl_inter_low
```
